```python
def nextMove(n,r,c,grid):
    pr = 0
    pc =0
    for i,row in enumerate(grid):
        if "p" in row:
            pr = i
            pc = row.index("p")
#     print(pr,pc)
    
    if pr > r:
        return "DOWN"
    elif pr < r:
        return "UP"
    elif pr==r:
        if(pc > c):
            return ("RIGHT")
        else:
            return("LEFT")
        
    return ""

n = int(input())
r,c = [int(i) for i in input().strip().split()]
# print(r,c)
grid = []
for i in range(0, n):
    grid.append(input())
# print(grid)
print(nextMove(n,r,c,grid))
```

```python

def next_move(posr, posc, board):
    up = posr - 1
    down = posr + 1
    right = posc + 1
    left = posc - 1
    
    if board[posr][posc] == "d":
        print("CLEAN")
        return
    
    if up>-1 and board[up][posc] == "d":
        print("UP")
        return

    if down<len(board) and board[down][posc] == "d":
        print("DOWN")
        return
    
    if right<len(board) and board[posr][right]=="d":
        print("RIGHT")
        return
    
    if left>-1 and board[posr][left]=="d":
        print("LEFT")
        return

    if "d" in board[posr] and board[posr][posc] != "d": 
        pos = board[posr].index("d")
        if pos > posc:
            print("RIGHT")
            return
        elif pos < posc:
            print("LEFT")
            return
  
    if up>-1 and "d" in board[up]:
        print("UP")
        return
    if down<len(board) and "d" in board[down]:
        print("DOWN")
        return
    
    for i, row in enumerate(board):
        if "d" in row and i< posr:
            print("UP")
            return
        elif "d" in row and i > posr:
            print("DOWN")
            return
            

# Tail starts here

if __name__ == "__main__":
    pos = [int(i) for i in input().strip().split()]
    board = [[j for j in input().strip()] for i in range(5)]
    next_move(pos[0], pos[1], board)
```

```python

#Using Manhattan distance

def getDist(item):
    return item.get("dist")
def next_move(posr, posc, board):
    manArray = []
    for i,row in enumerate(board):
        for j,ele in enumerate(row):
            
            if ele=="d":
                item = {}
                item["dist"] = abs(posr - i) + abs(posc-j)
                coordinates = []
                coordinates.append(i)
                coordinates.append(j)
                item["coord"] = coordinates
                manArray.append(item)
    manArray.sort(key=getDist)
    next = manArray[0]["coord"]
    
    if next[0] == posr:
        if next[1] > posc:
            return print("RIGHT")
        elif next[1] < posc:
            return print("LEFT")
        else:
            return print("CLEAN")
    elif next[0] > posr:
            return print("DOWN")
    else:
        return print("UP")
    
            

# Tail starts here

if __name__ == "__main__":
    pos = [int(i) for i in input().strip().split()]
    board = [[j for j in input().strip()] for i in range(5)]
    next_move(pos[0], pos[1], board)
```

```python
def get_dist(item):
    return item.get("dist")

def next_move(posx, posy, dimx, dimy, board):
    manht = []
    for i, row in enumerate(board):
        for j, ele in enumerate(row):
            if ele =="d":
                dist = abs(posx-i) + abs(posy-j)
                coord = []
                coord.append(i)
                coord.append(j)
                item= {}
                item["dist"] = dist
                item["coord"] = coord
                manht.append(item)
    manht.sort(key=get_dist)
    cell = manht[0]["coord"]
    x = cell[0]
    y = cell[1]
    if x == posx and y == posy:
        return print("CLEAN")
    elif x < posx:
        return print("UP")
    elif x > posx:
        return print("DOWN")
    elif y > posy:
        return print("RIGHT")
    else:
        return print("LEFT")
         

if __name__ == "__main__":
    pos = [int(i) for i in input().strip().split()]
    dim = [int(i) for i in input().strip().split()]
    board = [[j for j in input().strip()] for i in range(dim[0])]
    next_move(pos[0], pos[1], dim[0], dim[1], board)
```
